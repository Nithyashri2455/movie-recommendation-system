```python
import pandas as pd
a=pd.read_csv("movies.csv")
a
```

```python
a['genres'] = a['genres'].str.replace('|', ' ')
a['genres'] = a['genres'].str.lower()
a.head()
```

```python
a['genres'].value_counts().head()
```

```python
a = a[a['genres'] != '(no genres listed)']
```

```python
a=a.reset_index(drop=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
genre_matrix=cv.fit_transform(a['genres'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(genre_matrix)
similarity
```

```python
movie_indices = pd.Series(a.index, index=a['title']).drop_duplicates()
```

```python
def recommend(movie_title, n=5):
    idx = movie_indices[movie_title]
    sim_scores = list(enumerate(similarity[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:n+1]
    movie_indexes = [i[0] for i in sim_scores]
    return a['title'].iloc[movie_indexes]
```
